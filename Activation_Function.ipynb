# Demonstration of activation function

Activation Function code

```python
import numpy as np
import matplotlib.pyplot as plt
# Linear Activation
def linear(x):
  y=x
  return y

# signum activation
# let theta(threshold = 0)
def signum(x):
  if x==0:
    y=0
  elif x>0:
    y=1
  else:
    y=-1
  return y

# Sigmoid activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# tanh activation
def tanh(x):
    return np.tanh(x)

# ReLU activation
def relu(x):
    return np.maximum(0, x)

# Leaky ReLU activation
def leaky_relu(x):
    return np.maximum(0.1*x, x)

# ELU activation
def elu(x):
    return np.maximum(1.0*(np.exp(x)-1), x)
```

# Testing the various activation functions and plotting them

```python
#Testing for a range of -5 to 5
x=np.arange(-5,6,0.1,dtype=float)
print(x)
```

Linear Activation Function

```python
y=np.array([],dtype=float)
for i in x:
	y = np.concatenate([y, [linear(i)]])

fig, ax = plt.subplots()

# Set the position of the spines
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')


# Plot the data
ax.plot(x, y,label='linear')

# Add labels and legend
ax.set_title('Linear Activation Function')
ax.set_xlabel('Input')
ax.set_ylabel('Output')
ax.legend()

# Adjust the style of the plot
ax.grid(True)
ax.tick_params(axis='both', which='major', labelsize=10, direction='in', length=6, width=1)

# Set the size of the plot
fig.set_size_inches(6, 4)

# Show the plot
plt.show()
```
